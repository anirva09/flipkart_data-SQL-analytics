Analyzing Insights using Queries

 Revenue & Sales

```python
#What are the top 10 products by total revenue?
pd.read_sql("""
SELECT 
    p.product_name,
    SUM(o.final_price * o.units_sold) AS revenue
FROM orders o
JOIN products p ON o.product_id = p.product_id
GROUP BY p.product_name
ORDER BY revenue DESC
LIMIT 10
""", con=engine)
```

```python
#Which categories generate the highest revenue?
pd.read_sql("""
SELECT 
    p.category,
    SUM(o.final_price * o.units_sold) AS revenue
FROM orders o
JOIN products p ON o.product_id = p.product_id
GROUP BY p.category
ORDER BY revenue DESC
LIMIT 10;
""", con=engine)
```

```python
#what is the highest revenue by city?
pd.read_sql("""
SELECT 
    seller_city,
    SUM(final_price * units_sold) AS total_revenue
FROM analysis
GROUP BY seller_city
ORDER BY total_revenue DESC;
""", con=engine)
```

```python
#What is the average revenue per product/category?
pd.read_sql("""
SELECT 
    a.category,
    FORMAT(AVG(b.final_price * b.units_sold), 0) AS total_revenue
    FROM products a
    JOIN orders b ON
    a.product_id = b.product_id
    GROUP BY a.category
    ORDER BY total_revenue DESC
    LIMIT 5;
""", con=engine)
```

```python
#How does revenue vary over time (monthly trend)?
pd.read_sql("""
select 
    p.category,
    date_format(o.listing_date, '%%Y-%%m') as month,
    round(sum(o.units_sold * o.price), 0) as revenue
from orders o
join products p
    on o.product_id = p.product_id
group by date_format(o.listing_date, '%%Y-%%m'), p.category
order by month desc
limit 10;
""", con=engine)
```

Demand vs Supply

```python
#Which products have high demand but low stock?
pd.read_sql("""
SELECT 
    p.product_name,
    SUM(o.units_sold) AS total_units_sold,
    AVG(o.stock_available) AS avg_stock
FROM orders o
JOIN products p ON o.product_id = p.product_id
GROUP BY p.product_name
HAVING 
    SUM(o.units_sold) > 500   -- high demand threshold
    AND AVG(o.stock_available) < 50  -- low stock threshold
ORDER BY total_units_sold DESC
LIMIT 10;
""", con=engine)    
```

```python
#What is the demand-to-stock ratio for each product?
pd.read_sql("""
select 
    p.product_name as name,
    sum(o.units_sold) as total_demand,
    max(o.stock_available) as stock,
    round(sum(o.units_sold) / nullif(max(o.stock_available), 0), 2) as demand_to_stock
from orders o
join products p
    on p.product_id = o.product_id
group by p.product_name, p.product_id
order by demand_to_stock desc
limit 10;
""", con=engine)
```

```python
#Which categories are understocked or overstocked?
pd.read_sql("""
select 
    p.category,
    sum(o.units_sold) as total_demand,
    sum(o.stock_available) as total_stock,
    round(sum(o.units_sold) / nullif(sum(o.stock_available), 0), 2) as demand_to_stock_ratio,
    case 
        when sum(o.units_sold) > sum(o.stock_available) then 'understock'
        when sum(o.units_sold) < sum(o.stock_available) then 'overstock'
        else 'balanced'
    end as stock_status
from products p
join orders o
    on p.product_id = o.product_id
group by p.category
order by demand_to_stock_ratio desc;
""", con=engine)
```

```python
# Which categories are overstocked?
pd.read_sql("""
select 
    a.category,
    sum(b.units_sold) as demand,
    sum(b.stock_available) as stock,
    round(sum(b.units_sold) / sum(b.stock_available), 2) as demand_to_stock_ratio,
    case
        when sum(b.units_sold) < sum(b.stock_available) then 'overstock'
        else 'balanced'
    end as stock_status
from products a
join orders b
    on a.product_id = b.product_id
group by a.category,a.product_id
order by stock_status desc
limit 10;
""", con=engine)
```

Pricing & Discounts

```python
#What is the impact of discount_percent on units_sold?
pd.read_sql("""
select 
    case 
        when discount_percent between 0 and 10 then '0-10'
        when discount_percent between 11 and 20 then '11-20'
        when discount_percent between 21 and 30 then '21-30'
        else '30+'
    end as discount_bucket,
    avg(units_sold) as avg_units_sold
from orders o
join products p
    on o.product_id = p.product_id
group by discount_bucket
order by discount_bucket;
""", con=engine)
```

```python
#Do higher discounts lead to higher sales?
pd.read_sql("""
select 
    o.discount_percent,
    avg(o.units_sold) as avg_units_sold
from orders o
join products p
    on o.product_id = p.product_id
group by o.discount_percent
order by o.discount_percent;
    
""", con=engine)
```

```python
#Which products have high discounts but low sales?
pd.read_sql("""
select 
    a.product_name as name,
    sum(b.units_sold * b.price) as sales,
    avg(b.discount_percent) as discount_percent
from orders b
join products a
    on a.product_id = b.product_id
group by a.product_name, a.product_id
having avg(b.discount_percent) > 20
order by discount_percent desc,
sales asc
limit 10;
""", con=engine)
```

```python
#What is the average discount by category?
pd.read_sql("""
select 
    a.category,
    avg(b.discount_percent) as discount_percent
from orders b
join products a
    on a.product_id = b.product_id
group by a.category
order by discount_percent desc
limit 10;
""", con=engine)
```

Customer Behavior

```python
#What is the relationship between rating and units_sold?
pd.read_sql("""
select 
    r.rating,
    sum(o.units_sold) as total_units_sold
from orders o
join reviews r
    on o.product_id = r.product_id
group by r.rating
order by r.rating
limit 10;
""", con=engine)
```

```python
#Do products with more reviews sell more?
pd.read_sql("""
select 
    case 
        when r.review_count < 10 then 'low'
        when r.review_count between 10 and 50 then 'medium'
        else 'high'
    end as review_bucket,
    avg(o.units_sold) as avg_units_sold
from orders o
join reviews r
    on o.product_id = r.product_id
group by review_bucket
order by avg_units_sold;
""", con=engine)
```

```python
#Which products have high ratings but low sales?
pd.read_sql("""
select 
    p.product_name,
    avg(r.rating) as rating,
    sum(o.units_sold * o.price) as sales
from orders o
join products p on p.product_id = o.product_id
join reviews r on r.product_id = p.product_id
group by p.product_name, p.product_id
having avg(r.rating) >= 4
order by sales asc
limit 10;
""", con=engine)
```

```python
#Which products have low ratings but high sales?
pd.read_sql("""
select 
    p.product_name,
    avg(r.rating) as rating,
    sum(o.units_sold * o.price) as sales
from orders o
join products p on p.product_id = o.product_id
join reviews r on r.product_id = p.product_id
group by p.product_name, p.product_id
having avg(r.rating) < 3
order by sales desc
limit 10;
""", con=engine)
```

```python
#What is the average rating per category/brand?
pd.read_sql("""
select 
    p.brand,
    avg(r.rating) as avg_rating
from reviews r
join products p
    on p.product_id = r.product_id
group by p.brand
order by avg_rating desc;
""", con=engine)
```

Delivery Analysis

```python
#Does delivery_days affect units_sold?
pd.read_sql("""
select 
    l.delivery_days,
    avg(o.units_sold) as avg_units_sold
from orders o
join logistics l
    on o.product_id = l.product_id
group by l.delivery_days
order by l.delivery_days;
""", con=engine)
```

```python
#Which products have the fastest vs slowest delivery?
pd.read_sql("""
select 
    p.product_name,
    l.delivery_days
from logistics l
join products p
    on p.product_id = l.product_id
order by l.delivery_days asc
limit 10;
""", con=engine)
```

```python
#Which cities have longer delivery times?
pd.read_sql("""
select 
    s.seller_city,
    avg(l.delivery_days) as avg_delivery_days
from logistics l
join orders o
    on o.product_id = l.product_id
join sellers s
    on s.seller = o.seller
group by s.seller_city
order by avg_delivery_days desc;
""", con=engine)
```

```python
#Do faster deliveries lead to higher sales?
pd.read_sql("""
select 
    l.delivery_days,
    sum(o.units_sold) as total_units_sold
from orders o
join logistics l
    on o.product_id = l.product_id
group by l.delivery_days
order by l.delivery_days;
""", con=engine)
```

Returns & Policies

```python
#Do returnable products sell more than non-returnable ones?
pd.read_sql("""
select 
    r.is_returnable,
    sum(o.units_sold) as total_units_sold
from orders o
join returns r
    on o.product_id = r.product_id
group by r.is_returnable;
""", con=engine)
```

```python
#Which categories have higher return_policy_days?
pd.read_sql("""
select 
    p.category,
    avg(r.return_policy_days) as avg_return_days
from returns r
join products p
    on p.product_id = r.product_id
group by p.category
order by avg_return_days desc;
""", con=engine)
```

```python
#Is there a relationship between return_policy_days and sales?
pd.read_sql("""
select 
    r.return_policy_days,
    sum(o.units_sold) as total_units_sold
from orders o
join returns r
    on o.product_id = r.product_id
group by r.return_policy_days
order by r.return_policy_days;
""", con=engine)
```

```python
#Are highly sold products more likely to be returnable?
pd.read_sql("""
select 
    p.category,
    r.is_returnable,
    avg(t.total_units_sold) as avg_units_sold
from (
    select product_id, sum(units_sold) as total_units_sold
    from orders
    group by product_id
) t
join returns r on r.product_id = t.product_id
join products p on p.product_id = t.product_id
group by p.category, r.is_returnable
order by p.category;
""", con=engine)
```

Time-Based Analysis

```python
#How do newly listed products perform vs older ones?
pd.read_sql("""
select 
    case 
        when o.listing_date >= date_sub((select max(listing_date) from orders), interval 30 day) then 'new'
        when o.listing_date >= date_sub((select max(listing_date) from orders), interval 90 day) then 'mid'
        else 'old'
    end as product_age,
    avg(o.units_sold) as avg_units_sold,
    avg(o.units_sold * o.price) as avg_revenue
from orders o
group by product_age;
""", con=engine)
```

```python
#Is there a trend in product listings over time?
pd.read_sql("""
select 
    date_format(o.listing_date, '%%Y-%%m') as month,
    count(distinct o.product_id) as products_listed
from orders o
group by date_format(o.listing_date, '%%Y-%%m')
order by month;
""", con=engine)
```

```python
#Do older products have more reviews and sales?
pd.read_sql("""
select 
    case 
        when o.listing_date < date_sub((select max(listing_date) from orders), interval 90 day) then 'old'
        else 'new'
    end as product_age,
    avg(r.review_count) as avg_reviews,
    avg(o.units_sold) as avg_units_sold
from orders o
join reviews r
    on r.product_id = o.product_id
group by product_age;
""", con=engine)
```
